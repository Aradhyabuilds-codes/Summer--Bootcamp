Average of Students

```python
df['average_score'] = df[['math', 'science', 'english']].mean(axis=1)
print(df[['name', 'math', 'science', 'english', 'average_score']].head())
```

### Filter 1: Top Students (average score > 85)

```python
top_students = df[df['average_score'] > 85] # Filter students with an average score greater than 85
print('Students with average score greater than 85:')
display(top_students[['name', 'average_score', 'math', 'science', 'english']])
```

This table shows students who have an `average_score` greater than 85. These are considered the top performing students based on their core subject scores.

### Filter 2: Students with Low Attendance (attendance percent < 75)

```python
low_attendance_students = df[df['attendance_percent'] < 75] # Filter students with attendance percentage less than 75
print('Students with attendance percent less than 75:')
display(low_attendance_students[['name', 'attendance_percent', 'average_score']])
```

This table lists students with an `attendance_percent` less than 75. These students may require additional attention regarding their attendance.

### Filter 3: Students from Mathura

```python
mathura_students = df[df['city'] == 'Mathura'] # Filter students whose city is 'Mathura'
print('Students from Mathura:')
display(mathura_students[['name', 'city', 'average_score']])
```

# ***`STUDENT ANALYSER`***





### Create 'Result' Column (Pass/Fail)

```python
# Create 'result' column: Pass if average >= 40, otherwise Fail
df['result'] = df['average_score'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
print("DataFrame with 'result' column added (first 5 rows):")
display(df[['name', 'average_score', 'result']].head())
```

### Print Class Average

```python
# Print overall class average score
class_average = df['average_score'].mean()
print(f"Overall Class Average Score: {class_average:.2f}")
```

### Print Highest and Lowest Average Scores

```python
# Print highest and lowest average scores
highest_average = df['average_score'].max()
lowest_average = df['average_score'].min()
print(f"Highest Average Score: {highest_average:.2f}")
print(f"Lowest Average Score: {lowest_average:.2f}")
```

### Print Top Student(s)

```python
# Print student(s) with the highest average score
top_students_by_avg = df[df['average_score'] == df['average_score'].max()]
print("Top Student(s) by Average Score:")
display(top_students_by_avg[['name', 'average_score', 'math', 'science', 'english']])
```


Print students with attendance below 75.

```python
print('Students with attendance percent less than 75:')
display(low_attendance_students[['name', 'attendance_percent', 'average_score']])
```

### Print Students with Fewer Than 6 Assignments Submitted

```python
# Filter and print students who submitted fewer than 6 assignments
low_assignments_students = df[df['assignments_submitted'] < 6]
print("Students who submitted fewer than 6 assignments:")
display(low_assignments_students[['name', 'assignments_submitted', 'average_score']])
```

### Compare Average Marks of Mumbai Students and Pune Students

```python
# Check if 'Pune' exists in the 'city' column
if 'Pune' not in df['city'].unique():
    print("Note: There are no students from 'Pune' in the dataset. Comparing Mumbai students with available cities.")

mumbai_students_avg = df[df['city'] == 'Mumbai']['average_score'].mean()
print(f"Average score of Mumbai students: {mumbai_students_avg:.2f}")

# You can extend this to compare with other cities present in the dataset
available_cities = df['city'].unique()
other_cities_avg = {}
for city in available_cities:
    if city != 'Mumbai':
        other_cities_avg[city] = df[df['city'] == city]['average_score'].mean()
        print(f"Average score of {city} students: {other_cities_avg[city]:.2f}")
```

### Create 'performance_level' Column

```python
def assign_performance_level(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 75:
        return 'Good'
    elif score >= 60:
        return 'Average'
    else:
        return 'Poor'

df['performance_level'] = df['average_score'].apply(assign_performance_level)
print("DataFrame with 'performance_level' column added (first 5 rows):")
display(df[['name', 'average_score', 'performance_level']].head())
```
