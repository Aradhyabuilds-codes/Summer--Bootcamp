```python
students = [
    {
        "college_name": "gla",
        "name_stud": "aayush",
        "section": "cc",
        "city": "Matura",
        "Grades": "A+",
        "marks": 85
    },
    {
        "college_name": "gla",
        "name_stud": "akshay",
        "section": "cg",
        "city": "Matura",
        "Grades": "C+",
        "marks": 55
    },
    {
        "college_name": "gla",
        "name_stud": "Aman",
        "section": "cd",
        "city": "Matura",
        "Grades": "C+",
        "marks": 30
    },
    {
        "college_name": "gla",
        "name_stud": "ahmed",
        "section": "cc",
        "city": "Matura",
        "Grades": "B+",
        "marks": 70
    },
    {
        "college_name": "gla",
        "name_stud": "arnav",
        "section": "ca",
        "city": "Matura",
        "Grades": "A+",
        "marks": 92
    },
    {
        "college_name": "gla",
        "name_stud": "aham",
        "section": "cb",
        "city": "Matura",
        "Grades": "B+",
        "marks": 45
    }
]
```

```python
for student in students:
    student_name = student.get('name_stud')
    student_marks = student.get('marks')

    if student_marks > 80:
        performance = "Distinction"
    elif student_marks >= 40:
        performance = "Pass"
    else:
        performance = "Fail"

    print(f"{student_name}: {performance} (Marks: {student_marks})")
```

```python
a_plus_students = [student for student in students if student.get('Grades') == 'A+']
for student in a_plus_students:
    print(f"Name: {student.get('name_stud')}, Section: {student.get('section')}, Grades: {student.get('Grades')}")
```

```python
students = [
    {"name": "Palak", "marks": 92, "attendance": 95, "lab_assignment": True},
    {"name": "Akshat", "marks": 75, "attendance": 85, "lab_assignment": True},
    {"name": "Rishi", "marks": 58, "attendance": 60, "lab_assignment": True},
    {"name": "Aman", "marks": 88, "attendance": 90, "lab_assignment": False},
    {"name": "Ananya", "marks": 35, "attendance": 80, "lab_assignment": True}
]

for student in students:
    if student["lab_assignment"] == False:
        result = "Lab Assignment Not Submitted"
    elif student["attendance"] < 75:
        result = "Low Attendance"
    elif student["marks"] >= 70:
        result = "Distinction"
    elif student["marks"] >= 40:
        result = "Pass"
    else:
        result = "Fail"

    print(student["name"], "-", result)
```
