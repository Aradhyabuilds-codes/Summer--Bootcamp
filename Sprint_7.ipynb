```python
sentence = "the movie was not good"
tokens = sentence.split()
memory = ""

for token in tokens:
    memory = memory + " " + token
    print("Current token:", token)
    print("Memory so far:", memory.strip())
    print("-----")
```

```python
sentence = "the movie was not good"
tokens = sentence.split()

hidden_state = {
    "topic": None,
    "negation_active": False,
    "sentiment_word": None
}
for token in tokens:
    if token in ["movie", "food", "product"]:
        hidden_state["topic"]
    if token == "not":
        hidden_state["negation_active"] = True
    if token in ["good", "bad", "great", "terrible"]:
        hidden_state["sentiment_word"] = token

    print("Token:", token)
    print("Hidden state:", hidden_state)
    print("-----")
```

```python
import tensorflow as tf
from tensorflow.keras import layers

vocab_size = 1000
embedding_dim = 16
sequence_length = 8 # [cite: 79]

simple_rnn_model = tf.keras.Sequential([
    tf.keras.Input(shape=(sequence_length,)), # Explicitly define input shape
    layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim),
    layers.SimpleRNN(32),
    layers.Dense(1, activation="sigmoid")
])

simple_rnn_model.summary()
```
